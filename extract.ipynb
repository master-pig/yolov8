```python
video = cv2.VideoCapture("D:\资料\DAVE\DAVE THE DIVER 2024-02-14 00-39-01.mp4")
num = 0
save_step = 30
while True:
    ret , frame = video.read()
    if not ret:
        break
    num += 1
    if num % save_step == 0:
        cv2.imwrite("./demo_video/" + str(num) + ".jpg" , frame)
```
